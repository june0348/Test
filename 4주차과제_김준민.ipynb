### 심층 신경망

### MCP 뉴런
: 1943년 매컬럭과 월터 피츠는 신경세포를 간단한 바이너리 출력을 갖는 논리게이트로 설명한  맥컬럭-피츠(MCP)라 불리는 뇌 세포의 개념을 발표했다. 매컬럭과 월터 피츠가 발표한 MCP 세포는 우선 수상돌기에서 복수의 화학적 신호를 받아 세포체를 통해 통합해 전기 신호로 변경한다. 이후 전기신호가 임계치까지 누적된다면 축색돌기를 통해 다시 화학신호로 변경해 내보내는 간단한 구조를 갖고있었다. 

### 퍼셉트론
: MCP뉴런으로 부터 영감을 받은 프랭크 로젠블렛은 퍼셉트론이라는 임계치 모델과 학습 알고리즘을 발표했다. MCP뉴런의 논리게이트를 모델로 구현한것이기 때문에 근본적으로는 차이가 없다. 
퍼셉트론은 입력과 임계치를 포함한 활성함수, 출력으로 구성되있으며 입력값이 활성함수로 입력되게 된다. 이 때 입력값의 누적이 임계치(theta or threshold)보다 크면 1 을 반환하고 작으면 0을 반환 하는 바이너리 결과( 혹은 +1 과 -1) 를 출력값으로 갖는다.

### 역전파

### 강화학습

 에피소드(하나의 단계)가 끝나게 됐을 때, 에이전트가 지나왔던 상태에서 했던 행동에 대한 정보를 기록. 그리고 그 정보를 이용하여 다음 에피소드(다음 단계)에 대한 의사결정을 진행. 이와 같은 방법으로 이전에 얻어왔던 정보로 기록 업데이트 반복한다.

## 2. 딥러닝 강의 클론 코딩

####__퍼셉트론 구조 구현하기__ 
딥러닝 강의(__딥러닝 원리[1] 3:15 ~ 5:15 부분__)를 보고 코드를 따라 치며 출력 결과를 만드세요.
 

* 하나의 코드셀에 해당 코드를 한번에 다 적어서 실행해주세요 (__그렇게 하지 않을 경우, 아래 이미지와 같은 출력값이 나오지 않을 수 있습니다__)

*__주의!__ 실제로 코딩해서 출력해보면 강의에 나온 출력 결과와 다르게 나옵니다!!


![대체 텍스트](https://www.notion.so/image/https%3A%2F%2Fs3-us-west-2.amazonaws.com%2Fsecure.notion-static.com%2F0cceeed0-0235-4b0f-af88-0b8c377d5b4b%2F_2020-06-09__9.35.23.png?table=block&id=88fd8912-9356-49a4-9fda-a1a63fe96ea9&width=2870&cache=v2)

```
# 입력값이 1, 원하는 출력값 0
# 입력x --(가중치w)--> 활성화 함수f(x) --> 출력y
# y=f(x*w)

import tensorflow as tf
tf.compat.v1.set_random_seed(2020)
x = 1                                                                           # 입력값 1
y = 0                                                                           # 원하는 출력값 0
w=tf.random.normal([1], 0, 1)                                                   # 가중치 정규분포의 무작위 값 설장

import math                                                                     # 활성화 함수 생성
def sigmoid(x) :
  return 1/(1 + math.exp(-x))

# 입력값과 가중치가 곱해저 활성화 함수를 통과 할때 출력값을 확인
output = sigmoid(x*w)
print(output)

# 0과 먼 이유는 가중치를 무작위로 선청했고, 이 가중치는 원하는 출력값을 얻기에는 적절한 가중치가 아님을 의미

# error = 0(기대출력) - 0.474771(실제출력)
# 가중치 조정을 1000번 조정

for i in range(1000):
  output = sigmoid(x*w)
  error = y - output
  w = w + x*0.1*error                                                           # 경사하강법  w = w + x*n(학습률)*error/ 학습률: 가중치 조정을 위한 하이퍼파라미터

  if i % 100 == 99:
    print("학습횟수:", i, "Error:", error, "예측결과:", output)
```
